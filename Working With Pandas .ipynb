```python
columns='Yr Mo Dy RPT VAL ROS KIL SHA BIR DUB CLA MUL CLO BEL MAL'  #Method -1
columns=columns.split()
data=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data',header=None)
data=data.set_axis(['data'],axis=1)
data=data['data'].str.split(expand=True).set_axis(columns,axis=1)
```

```python
data=data.iloc[1:,:]
```

```python
data[['RPT','VAL','ROS','KIL','SHA','BIR','DUB','CLA','MUL','CLO','BEL','MAL']]=data[['RPT','VAL','ROS','KIL','SHA','BIR','DUB','CLA','MUL','CLO','BEL','MAL']].astype('float')
data[['Yr','Mo','Dy']]=data[['Yr','Mo','Dy']].astype('int')
```

```python
# Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index
# Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
date_cols = {
    'year': data['Yr'] + 1900,
    'month': data['Mo'],
    'day': data['Dy'] }

data.insert(0,'Date',pd.to_datetime(date_cols))
data.drop(["Yr","Mo","Dy"],axis=1)
data=data.reset_index(drop=True)
```

```python
# Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
data.set_index("Date")
```

```python
# Step 6. Compute how many values are missing for each location over the entire record.
missing=data.isna().sum().sum()
missing
```

```python
# Step 7. Compute how many non-missing values there are in
len(data)-missing
```

```python
# Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
data.mean().mean()
```

```python
# Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of
#         the windspeeds at each location over all the days
loc_stats = data.describe().loc[["min","max","mean","std"]]
loc_stats
```

```python
# Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the
#          windspeeds across all the locations at each day.
day_stats=data.groupby("day").describe().loc[["min","max","mean","std"]]
day_stats
```

```python
data.groupby('Dy')[['BIR', 'RPT']].agg(['min', 'max', 'std', 'mean'])
```

```python
# Step 11. Find the average windspeed in January for each location.
year=data[data["Date"].dt.year.isin([1961,1962])]
year.groupby(year["Date"].dt.month ==1).mean().iloc[1]
```
